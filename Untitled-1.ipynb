```python
import os
import random
import shutil


# Set the paths for the original dataset and the destination for the impostor dataset
original_dataset_path = "/Users/stellafazioli/Documents/Casia-evalu/train"
impostor_dataset_path = "/Users/stellafazioli/Documents/Casia-evalu/impostor/"
# Create the impostor dataset directory structure
if not os.path.exists(impostor_dataset_path):
    os.makedirs(impostor_dataset_path)

# Get a list of subfolders (classes) in the original dataset
classes = [subfolder for subfolder in os.listdir(original_dataset_path) if os.path.isdir(os.path.join(original_dataset_path, subfolder))]

# Iterate over each subfolder (class) in the original dataset
for cls in classes:
    class_folder = os.path.join(original_dataset_path, cls)
    impostor_class_folder = os.path.join(impostor_dataset_path, cls)

    # Create the impostor subfolder for the current class
    if not os.path.exists(impostor_class_folder):
        os.makedirs(impostor_class_folder)

    # Get a list of image files in the current class folder
    image_files = [file for file in os.listdir(class_folder) if file.endswith(".jpg") or file.endswith(".png")]

    # Copy true images to the impostor class folder
    for img in image_files:
        true_image_path = os.path.join(class_folder, img)
        true_image_dest = os.path.join(impostor_class_folder, img)
        shutil.copyfile(true_image_path, true_image_dest)

    # Copy impostor images from other classes to the impostor class folder
    for img in image_files:
        if img not in image_files:
            impostor_class = random.choice(classes)
            while impostor_class == cls:
                impostor_class = random.choice(classes)

            impostor_image_path = os.path.join(original_dataset_path, impostor_class, random.choice(os.listdir(os.path.join(original_dataset_path, impostor_class))))
            impostor_image_dest = os.path.join(impostor_class_folder, f"{impostor_class}_{os.path.basename(impostor_image_path)}")
            shutil.copyfile(impostor_image_path, impostor_image_dest)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

val_dir="/Users/stellafazioli/Downloads/archive-2/lfw-deepfunneled/lfw-deepfunneled/"
image_size = (150, 150)

# Set the number of classes (number of individuals in the dataset)
num_classes = 5749

# Set the batch size and number of epochs for training
batch_size = 32
num_epochs = 10

data_generator = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)


val_generator = data_generator.flow_from_directory(
    val_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
import dlib
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Load the pre-trained FaceNet model from dlib
model_path = '/Users/stellafazioli/Downloads/dlib_face_recognition_resnet_model_v1.dat'
facenet_model = dlib.face_recognition_model_v1(model_path)
```

```python




num_samples = len(val_generator)
binary_labels = val_generator.labels

# Compute the face embeddings for each face image from the generator
face_embeddings = []
for i in range(num_samples):
    face_images, _ = val_generator[i]  # Get batch of face images from the generator
    for face_image in face_images:
        # Convert face_image to RGB if needed
        if face_image.shape[2] == 1:
            face_image = np.repeat(face_image, 3, axis=2)
        elif face_image.shape[2] == 4:
            face_image = face_image[:, :, :3]
        # Resize face_image to match the input size of the FaceNet model
        face_image = cv2.resize(face_image, (64, 64))
        # Compute the face descriptor
        face_descriptor = facenet_model.compute_face_descriptor(face_image)
        face_embeddings.append(np.array(face_descriptor))

# Compute the similarity scores between face embeddings
scores = []
for i in range(len(face_embeddings)):
    for j in range(i + 1, len(face_embeddings)):
        distance = np.linalg.norm(face_embeddings[i] - face_embeddings[j])
        similarity_score = 1 / (1 + distance)  # Similarity score between 0 and 1
        scores.append(similarity_score)


# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds
fpr, tpr, thresholds = roc_curve(binary_labels, scores)

# Compute the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Random guessing line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
